```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('./data/general/cleaned_fakenewsdata.csv')
print("Dataset shape:", df.shape)
df.head()
```

```python
df['label'].value_counts()
```

```python
X, y = df['text'], df['label']
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
tfidf = vectorizer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(tfidf, y, test_size=0.2, random_state=42)
X_train_t, X_val, y_train_t, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
model = SGDClassifier(loss='hinge', max_iter=1000, tol=1e-3, random_state=42)
model.fit(X_train_t, y_train_t)
y_pred = model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_pred))
print(classification_report(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
```
